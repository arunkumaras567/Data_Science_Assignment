```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
q9_a=pd.read_csv(r'C:/Users/admin/OneDrive/Desktop/Q9_a.csv',index_col='Index')
```

```python
q9_a.head()
q9_a.tail()
```

```python
print('For cars speed:',"skewness value=", np.round(q9_a.speed.skew(),2),'and','kurtosis value=',np.round(q9_a.speed.kurt(),2))
```

```python
print('For cars distance:',"skewness value=", np.round(q9_a.dist.skew(),2),'and','kurtosis value=',np.round(q9_a.dist.kurt(),2))
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
sns.distplot(q9_a.speed)
plt.subplot(1,3,3)
sns.distplot(q9_a.dist)
```

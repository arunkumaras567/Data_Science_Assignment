# Compare Evidences with Hypothesis using t-statistics

```python
statistics, p_value=st.ttest_ind(Cutlets['Unit A'],Cutlets['Unit B'], alternative='two-sided')
print('p_value:',p_value)
```

```python
Alpha=0.025
print('Significance=%.3f , p_value=%.3f' % (Alpha,p_value))
if p_value <= Alpha:
    print('We reject Null Hypothesis there is a significant difference between two Units A and B')
else:
    print('We fail to reject Null hypothesis')
```

# Hence, We fail to reject Null Hypothesis because of lack of evidence, there is no significant difference between the two samples.

# Question 3

We are going to conduct a Test of Independence using Chi-Square test with Contingency table

We need to check whether the proportion of any of these samples are different or the same?

```python
buyer_ratio=pd.read_csv(r'C:/Users/admin/OneDrive/Desktop/BuyerRatio.csv',index_col=0)
```

# collect evidence importing file

```python
buyer_ratio
```

```python
table=[[50,142,131,70],
      [435,1523,1356,750]] 
```

Applying Chi-Square contingency table to convert observed value into expected value

```python
st.chi2_contingency(table) 
```

```python
st.chi2_contingency(buyer_ratio)
```

```python
observed=np.array([50,142,131,70,435,1523,1356,750])
expected=np.array([ 42.76531299,  146.81287862,  131.11756787,   72.30424052, 442.23468701, 1518.18712138, 1355.88243213,  747.69575948])
```

# comparing evidence with hypothesis

```python
statistics, p_value = st.chisquare(observed,expected, ddof = 3)
print("Statistics = ",statistics,"\n",'P_Value = ', p_value)
```

```python
alpha = 0.05
print('Significnace=%.3f, p=%.3f' % (alpha, p_value))
if p_value <= alpha:
    print('We reject Null Hypothesis there is a association or dependency between Male and Female buyers ratios are similar across all regions')
else:
    print('We fail to reject Null hypothesis')
```

# We fail to reject Null Hypothesis because of lack of evidence. Therefore, there is no association or dependency between male-female buyers ratios and are similar across regions. Hence, Independent samples.

```python
statistics, p_value = st.chisquare(observed,expected, ddof = 3)
print("Statistics = ",statistics,"\n",'P_Value = ', p_value)
```

```python
alpha = 0.05
print('Significnace=%.3f, p=%.3f' % (alpha, p_value))
if p_value <= alpha:
    print('We reject Null Hypothesis the defective % didnot varies by centre')
else:
    print('We fail to reject Null hypothesis')
```
